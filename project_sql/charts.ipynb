### LEADING COUNTRIES FOR JOB POSTINGS

```python
import pandas as pd

df = pd.read_csv(r"D:\Downloads\leading_countries.csv")
df.head(10)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

world = gpd.read_file(r"D:\Downloads\110m_cultural\ne_110m_admin_0_countries.shp")
country_replacements = {
    "United States": "United States of America",
    "Russia": "Russian Federation",
    "South Korea": "Republic of Korea",
    "North Korea": "Dem. Rep. Korea",
    "Iran": "Iran (Islamic Republic of)",
    "Vietnam": "Viet Nam",
    "Czech Republic": "Czechia",
    "Syria": "Syrian Arab Republic",
    "Laos": "Lao PDR",
    "Bolivia": "Bolivia (Plurinational State of)",
    "Venezuela": "Venezuela (Bolivarian Republic of)",
    "Tanzania": "United Republic of Tanzania",
    "Sudan": "Sudan",  # in your case, already fine
    "Congo (Kinshasa)": "Democratic Republic of the Congo",
    "Congo (Brazzaville)": "Republic of the Congo"
}

df["job_country"] = df["job_country"].replace(country_replacements)

# Merge
merged = world.merge(df, how="left", left_on="ADMIN", right_on="job_country")

# Plot the map
fig, ax = plt.subplots(figsize=(12, 6), facecolor="white")
merged.plot(ax=ax, color="white", linewidth=0.3, edgecolor="black")

# Choropleth overlay
plot = merged.plot(
    column="job_count",
    cmap="Blues",
    linewidth=0.4,
    ax=ax,
    edgecolor="black",
    legend=True,
    missing_kwds={"color": "lightgrey"}
)

# Title
ax.set_title(
    "Global Analyst Job Postings",
    fontdict={"fontsize": 16, "color": "black", "weight": "bold"},
    pad=15
)

# Grid lines
ax.grid(color="white", linestyle="--", linewidth=0.3)

# Legend
cbar = ax.get_figure().get_axes()[1]
cbar.tick_params(labelsize=9, colors="black")
cbar.set_ylabel("Job Postings", color="black", fontsize=10)

plt.tight_layout()
plt.show()
```

### 2. TOP SKILLS FOR DATA ANALYST
#### Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df3 = pd.read_csv('D:\Downloads\Top_skills_for_DA.csv')
df3
```

```python
skill_counts = df3["skill_name"].value_counts().head(10)
skill_counts = skill_counts.sort_values(ascending=True)
skills = skill_counts.index
counts = skill_counts.values
```

### Bar chart

```python
colors = plt.cm.Blues(np.linspace(0.4, 1, len(skills)))

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(skills, counts, color=colors)

plt.title("Top 10 Skills for Data Analyst", fontsize=14, weight="bold")
plt.xlabel("Frequency")
plt.ylabel("Skills")

for bar in bars:
    height = bar.get_height()
    
plt.tight_layout()
plt.show()
```
