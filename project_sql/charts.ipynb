## TOP SKILLS FOR DATA ANALYST

### Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('D:\Downloads\Top_skills_for_DA.csv')
df
```

```python
skill_counts = df["skill_name"].value_counts().head(10)
skill_counts = skill_counts.sort_values(ascending=True)
skills = skill_counts.index
counts = skill_counts.values
```

### Bar chart

```python
colors = plt.cm.Blues(np.linspace(0.4, 1, len(skills)))

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(skills, counts, color=colors)

plt.title("Top 10 Skills for Data Analyst", fontsize=14, weight="bold")
plt.xlabel("Frequency")
plt.ylabel("Skills")

for bar in bars:
    height = bar.get_height()
    
plt.tight_layout()
plt.show()
```
