```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("D:\Downloads\Tutorial5Ex4.csv")

# Define the custom color palette (Blue, Orange, Green)
custom_palette = {
    'setosa': 'blue',
    'versicolor': 'orange',
    'virginica': 'green'
}

# --- Create the scatter plot with all circle markers ---
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='Sepal.Length',
    y='Sepal.Width',
    hue='Species',         # Map Species to color
    marker='o',            # Set all markers to 'o' (circle)
    s=100,                 # Increase marker size
    palette=custom_palette # Apply the custom color palette
)

# Add title and labels
plt.title('Sepal Length vs. Sepal Width by Species', fontsize=16)
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Sepal Width (cm)', fontsize=12)

# Move the legend outside the plot for better clarity
legend = plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("D:\Downloads\Tutorial5Ex4.csv")

# Define the custom color palette (Blue, Orange, Green)
custom_palette = {
    'setosa': 'blue',
    'versicolor': 'orange',
    'virginica': 'green'
}

# --- Create the Pair Plot (Pair Grid) ---
# This generates scatter plots for all pairs of numerical columns 
# and histograms/KDEs on the diagonal.
plt.figure(figsize=(10, 10)) # Note: pairplot creates its own figure, but it's good practice
                               # to set a size for clarity/preview in some environments.

# 'hue' maps the categorical 'Species' column to the color of the points.
# 'palette' applies the custom colors.
# 'diag_kind' specifies the type of plot for the diagonal (e.g., 'hist' or 'kde').
pair_plot = sns.pairplot(
    df, 
    hue='Species', 
    palette=custom_palette, 
    diag_kind='kde' # Use Kernel Density Estimate for the diagonal plots
)

# Optional: Add a main title for the figure
pair_plot.fig.suptitle('Pairwise Relationships of Iris Features by Species', y=1.02, fontsize=16)

# Show the plot
plt.show()
```

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt 

# Load the dataset
df = pd.read_csv("D:\Downloads\Tutorial5Ex6.csv")

# --- Data Preparation ---
# Assuming the last column is the categorical one for coloring (e.g., 'Category')
# You may need to replace 'Category' with the actual name of your categorical column.
categorical_column_name = df.columns[-1]

# Identify the numerical columns to be used as parallel axes
# This selects all columns EXCEPT the last one
numerical_features = df.columns[:-1]

# --- Create the Parallel Coordinate Plot ---
# This plot maps each row (data point) as a line, with its coordinates determined 
# by the values on the respective feature axes.
fig = px.parallel_coordinates(
    df,
    # The dimensions argument specifies which columns become the parallel axes
    dimensions=numerical_features,
    # The color argument maps a categorical column to the line color
    color=categorical_column_name,
    color_continuous_scale=px.colors.diverging.Tealrose, # Customize color scheme
    title="Parallel Coordinate Plot of Data Features by Category"
)

# Display the interactive plot
fig.show()
```
